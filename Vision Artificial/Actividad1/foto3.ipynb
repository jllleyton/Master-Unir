```python
import cv2
import numpy as np

# Leer imagen nocturna
img = cv2.imread("foto3_original.png")

# 1. Denoising
den = cv2.fastNlMeansDenoisingColored(img, None, 10, 10, 7, 21)

# 2. Gamma correction
gamma = 0.7
img_gamma = np.power(den / 255.0, gamma)
img_gamma = np.uint8(img_gamma * 255)

# 3. CLAHE en luminancia
lab = cv2.cvtColor(img_gamma, cv2.COLOR_BGR2LAB)
l, a, b = cv2.split(lab)

clahe = cv2.createCLAHE(clipLimit=1.8, tileGridSize=(8,8))
l_clahe = clahe.apply(l)

lab = cv2.merge((l_clahe, a, b))
final = cv2.cvtColor(lab, cv2.COLOR_LAB2BGR)

# 4. Normalizaci√≥n
final = cv2.normalize(final, None, 0, 255, cv2.NORM_MINMAX)

cv2.imwrite("foto3_procesada.png", final)
```
