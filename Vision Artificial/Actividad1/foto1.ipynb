```python
#foreground
#Denoising → Gamma (0.6–0.8) → CLAHE → Estiramiento no lineal
import cv2
import numpy as np

img = cv2.imread("foto1_original.png")

# --- Pipeline previo ---
#den = cv2.fastNlMeansDenoisingColored(img, None, 10, 10, 7, 21)

gamma = 0.3
img_gamma = np.power(img / 255.0, gamma)
img_gamma = np.uint8(img_gamma * 255)

# 3. CLAHE en luminancia
lab = cv2.cvtColor(img_gamma, cv2.COLOR_BGR2LAB)
l, a, b = cv2.split(lab)

clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
l_clahe = clahe.apply(l)

# --- Estiramiento no lineal (sigmoide) ---
l_norm = l_clahe / 255.0
k = 10      # fuerza
x0 = 0.45    # centro (primer plano)

l_nonlinear = 1 / (1 + np.exp(-k * (l_norm - x0)))
l_nonlinear = np.uint8(l_nonlinear * 255)

lab_final = cv2.merge((l_nonlinear, a, b))
final = cv2.cvtColor(lab_final, cv2.COLOR_LAB2BGR)

cv2.imwrite("foto1_procesada.png", final)
```
